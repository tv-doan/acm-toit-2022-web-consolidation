```python
import pandas as pd
```

```python
com = pd.read_csv('com_by_asn.csv')
net = pd.read_csv('net_by_asn.csv')
org = pd.read_csv('org_by_asn.csv')
```

```python
com = com.drop(com.filter(regex='apex').columns, axis=1)
net = net.drop(net.filter(regex='apex').columns, axis=1)
org = org.drop(org.filter(regex='apex').columns, axis=1)
```

```python
com.columns = ['date'] + list(map(lambda x: x[4:], com.columns[1:]))
```

```python
net.columns = ['date'] + list(map(lambda x: x[4:], net.columns[1:]))
```

```python
org.columns = ['date'] + list(map(lambda x: x[4:], org.columns[1:]))
```

```python
asns_cdns = pd.read_csv('./asn_cdns.csv', sep=';')
```

```python
asns_cdns.drop_duplicates(['cdn_name', 'asn'], keep='first', inplace=True)
```

```python
cdns = asns_cdns[['cdn_name', 'asn']].set_index('asn')
```

```python
cdns.index = cdns.index.astype(str, copy = False)
```

```python
cdns.index
```

```python
dfs_new = []
```

```python
for df in [com, net, org]:
    df = df.transpose().join(cdns, how='left')
    df = df.groupby('cdn_name').sum().append(df.iloc[-1]).transpose().set_index('date').dropna()
    dfs_new.append(df)
```

```python
dfs_new[0].to_csv('com_by_asn_processed.csv')
dfs_new[0]
```
